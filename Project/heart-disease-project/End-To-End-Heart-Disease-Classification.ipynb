# Simple Heuristic for Heart Disease Prediction

We can create a basic heuristic based on gender to predict the presence of heart disease:

- **Women:**
  - Total: 100
  - With heart disease: 72
  - **Probability:** 72/100 = 75%

- **Men:**
  - Total: 200
  - With heart disease: 100 (approximately)
  - **Probability:** 100/200 = 50%

By averaging these probabilities:
- **Overall Probability:** (75% + 50%) / 2 = 62.5%

### Conclusion
Based on this heuristic, we can estimate a 62.5% chance that any person has heart disease. This simple model serves as our baseline, which we aim to improve using machine learning techniques.
