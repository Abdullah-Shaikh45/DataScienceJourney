### What is Seaborn?

Seaborn is a Python data visualization library built on top of Matplotlib. It provides a high-level interface for creating <span style="color:red">attractive</span> and <span style="color:red">informative</span> statistical graphics.

### Why Use Seaborn?

1. **Simplifies Complex Plots:** Seaborn makes it easier to create complex visualizations with fewer lines of code.
2. **Built-in Themes:** It offers <span style="color:red">attractive</span> default styles and color palettes to make plots more visually appealing.
3. **Statistical Plots:** Seaborn includes functions for creating commonly used statistical plots like bar plots, box plots, heatmaps, and scatter plots.
4. **Integration with Pandas:** It works seamlessly with pandas DataFrames, allowing for easy plotting of data directly from these structures.

### Why Use Seaborn When You Have Matplotlib?

While it's true that Matplotlib can accomplish everything Seaborn does, Seaborn provides a more <span style="color:red">user-friendly</span> and <span style="color:red">efficient</span> way to generate <span style="color:red">beautiful</span> and <span style="color:red">informative</span> plots with less code. This makes it particularly useful for:

#### Quick Prototyping
When you need to quickly visualize data without spending much time on customization.

#### Statistical Analysis
When you require specialized plots that are commonly used in statistical data analysis.

#### Aesthetic Consistency
When you want consistent and attractive styling without manually tweaking every element.

In summary, Seaborn enhances your data visualization workflow by offering <span style="color:red">simplicity</span>, <span style="color:red">beautiful defaults</span>, and <span style="color:red">specialized statistical plotting capabilities</span>, making it an excellent complement to Matplotlib.


```python
# Basic import

import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Import Data

```python
# Builtin Datasets (choose any of these)

print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

## Distribution plot:
- A Distplot or distribution plot, depicts the variation in the data distribution.

```python
# Let's say we want to see the (Distribution Plot) for ['not_distracted']

sns.distplot(crash_df['not_distracted']);
```

```python
# We can also add (bins, kde)
# bins: How distributed value will be
# kde: kernel density estimate (KDE) 
# represents the data using a continuous probability density curve in one or more dimensions

sns.distplot(crash_df['not_distracted'], kde=False, bins=20);
```

## Joint plot: 
- A univariate plot that shows the comparison of different groups in categorical variables. 
- It shows the number of observations per category using bins.

```python
# Now we'll use Speed and Alcohol
# kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }

sns.jointplot(data=crash_df, x='speeding', y='alcohol', kind='reg');
```

## KDE plot

```python
sns.kdeplot(crash_df['alcohol']);
```

# Pair plots:
- The Seaborn Pairplot allows us to plot pairwise relationships between variables within a data set.

```python
sns.pairplot(crash_df);
```

```python
# Loading another dataset
# hue:  hue is a parameter that allows you to split and color

tips_df = sns.load_dataset('tips')
sns.pairplot(data=tips_df, hue='sex', palette="Greens");
```

## Rug plots

```python
sns.rugplot(tips_df['tip']);
```

## Styling: Your plots

```python
sns.set_style('darkgrid')

plt.figure(figsize=(8, 4))

sns.set_context('paper', font_scale=1.4)

sns.jointplot(data=crash_df, x='speeding', y='alcohol', kind='reg');
```

# Categorical Plots

### Bar plots

```python
sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.mean, color="salmon");
```

```python
# You add different color using paletts

sns.barplot(x='sex', y='total_bill', data=tips_df, estimator=np.mean, palette=["salmon", "Pink"])
```

### Countplot: 
- Counts the no. of occurences

```python
sns.countplot(x='sex', data=tips_df, palette=['salmon', 'lightyellow']);
```

## Box Plot:
- A box plot in Seaborn visually represents the distribution of a dataset through its quartiles, highlighting the median, interquartile range, and potential outliers.

```python
sns.boxplot(x='day', y='total_bill', data=tips_df, hue='sex')

plt.legend(loc=0);
```

## Violin Plot: 

- A violin plot is a type of graphical representation used to visualize the distribution of numeric data across different categories. 
- It combines aspects of a box plot and a density plot.

```python
tips_df.head()
```

```python
sns.violinplot(data=tips_df, x='day', y='total_bill', hue='sex');
```

```python
# We can also add split here (split=True) Which is used for comparison
# hue:  hue is a parameter that allows you to split and color

sns.violinplot(data=tips_df, x='day', y='total_bill', hue='sex', split=True);
```

# Strip plot:
- Stripplot is another type of categorical scatter plot where all data points are shown along the categorical axis (x or y), potentially overlapping.

```python
# Customizing figure

plt.figure(figsize=(10, 8))

# Jitter (jitter=True): Prevents overlap of points with the same data values.
# Dodge (dodge=True): Separates points by hue categories along the categorical axis.

sns.stripplot(data=tips_df, x='day', y='total_bill', hue='sex', jitter=True, dodge=True);
```

## Swarm-plots: 
Swarmplot is a type of categorical scatter plot where points are plotted in a way that avoids overlapping, typically along one categorical axis (x or y).

```python
sns.swarmplot(data=tips_df, x='day', y='total_bill', hue='sex');
```

# Customization of plots

```python
# We'll take an example here

# 1. Making this figure big
plt.figure(figsize=(10, 8))

# 2. Setting style {darkgrid, whitegrid, dark, white, ticks}
sns.set_style('darkgrid')

# 3. Context: These parameters correspond to label size, line thickness, etc {paper, notebook, talk, poster}
sns.set_context('notebook')

# 0. Plot
sns.stripplot(data=tips_df, x='day', y='total_bill', hue='sex', palette='magma')

# 4. Legend: Show info about the plots (0, 1, 2, 3, 4)
plt.legend(loc=2);
```

# Matrix Plots

### Heatmap Visualization

#### Overview
- **Heatmap:** A graphical representation of data where values in a matrix are represented as colors. It's **useful** for **visualizing patterns** and **trends**, especially in large datasets, by using **color intensity** to show variations in values.


```python
plt.figure(figsize=(8, 6))
sns.set_context('paper', font_scale=1.4)

# Exclude non-numeric columns before computing correlation matrix
numeric_cols = crash_df.select_dtypes(include=['float64', 'int64']).columns
crash_mx = crash_df[numeric_cols].corr()
crash_mx

# Plot heatmap
sns.heatmap(crash_mx, annot=True, cmap='Blues')
```

```python
# Customization
plt.figure(figsize=(10, 8))
sns.set_context('paper', font_scale=1.4)
plt.title('Correlation Heatmap of Numeric Columns')

# Excluding non-numeric columns
numeric_cols = crash_df.select_dtypes(include=['float64', 'int64']).columns
crash_mat = crash_df[numeric_cols].corr()

# Plotting Heatmap ((annot=True) display data values in each heatmap cell & `cmap`= color-map)
sns.heatmap(data=crash_mat, annot=True, cmap='Blues');
```

```python
# Another Example

# Importing "Flights-Datasets"
flights_df = sns.load_dataset('flights')

# Using `pivot_table()`
flights_df = flights_df.pivot_table(index='month', columns='year', values='passengers')

# Plotting
sns.heatmap(flights_df, cmap='Blues', linecolor='white', linewidths=1);
```

### Cluster Map

```python
sns.clustermap(flights_df, cmap='Blues', standard_scale=1);
```

### Pair Grid

```python
iris = sns.load_dataset('iris')
iris_g = sns.PairGrid(iris, hue='species')
iris_g.map(plt.scatter);
```

```python
# Load the iris dataset
iris = sns.load_dataset('iris')

# Create a PairGrid object with specified variables
iris_g = sns.PairGrid(iris, hue='species', 
                      x_vars=['sepal_length', 'sepal_width'], 
                      y_vars=['petal_length', 'petal_width'])

# Map the diagonal with histograms
iris_g.map_diag(plt.hist)

# Map the upper triangle with scatter plots
iris_g.map_upper(plt.scatter)

# Map the lower triangle with kernel density plots
iris_g.map_lower(sns.kdeplot)

# Add a legend
iris_g.add_legend()

# Show the plot
plt.show()
```

### Facet Grid:
- A FacetGrid in Seaborn is a way to create a grid of plots based on the values of one or more categorical variables. It allows you to visualize the relationship between a pair of variables across different subsets of the data.

```python
# tips_fg: FacetGrid
tips_fg = sns.FacetGrid(tips_df, col='time', row='smoker')

# 1. Adding additional things & Trying out
tips_fg.map(plt.hist, 'total_bill', bins=8);
```

```python
# 2. Adding additional things & Trying out

tips_fg.map(plt.scatter, 'total_bill', 'tip');
```

```python
# 3. Adding additional things & Trying out
tips_fg = sns.FacetGrid(tips_df, col='time', hue='smoker', height=4, aspect=1.3, 
                        col_order=['Dinner', 'Lunch'], palette='Set1')


tips_fg.map(plt.scatter, 'total_bill', 'tip', edgecolor='w');
```
