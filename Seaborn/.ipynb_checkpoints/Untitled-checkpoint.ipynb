# Seaborn revesion

```python
# Basic import 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Data

```python
# Built-in Datasets (choose any one of it)

print(sns.get_dataset_names())
```

```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head()
```

```python
crash_df.dtypes
```

### Distribution Plot

```python
sns.distplot(crash_df['not_distracted'])
```

```python
# We can also add (bins, kde)
# `bins`: How distributed the value will be
# `kde`: Kernel Density Estimate (Probality of density curve)

sns.distplot(crash_df['not_distracted'], kde=False, bins=20);
```

# Joint Plot

```python
# kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }

sns.jointplot(data=crash_df, x='speeding', y='alcohol', kind='reg')
```

### KDE Plot

```python
sns.kdeplot(crash_df['speeding']);
```

### Pair Plots 

```python
sns.pairplot(crash_df);
```

```python
# Loading Another datasets
# hue: It is used to allow and split color

tips_df = sns.load_dataset('tips')

# Using pairplot
sns.pairplot(data=tips_df, hue='sex', palette='Blues');
```

```python
## Rug Plot

sns.rugplot(tips_df);
```
